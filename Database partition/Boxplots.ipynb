# Boxplots function

```python
#Brayan Arteaga - UAM
def bigotes(CNN_name,Steg_name,A,B,C,D):
    with plt.style.context('seaborn-white'):
        plt.ion()
        plt.boxplot([A, B, C, D], sym = 'ko', whis = 1.5)
        plt.xticks([1,2,3,4], ['A', 'B', 'C', 'D'])
        plt.ylabel('Accuracy')
        plt.grid()
        plt.title(CNN_name)
        plt.tight_layout(pad=3.0)
```
